```python
import pandas as pd
from scipy.fftpack import idct, dct
import numpy as np
import numpy as np
import pylab as pl
from numpy import fft
import matplotlib.pyplot as plt


def fourierExtrapolation(x, n_predict):
    n = x.size
    n_harm = 10                     # number of harmonics in model
    t = np.arange(0, n)
    p = np.polyfit(t, x, 1)         # find linear trend in x
    x_notrend = x - p[0] * t        # detrended x
    x_freqdom = fft.fft(x_notrend)  # detrended x in frequency domain
    f = fft.fftfreq(n)              # frequencies
    indexes = list(range(n))
    # sort indexes by frequency, lower -> higher
    indexes.sort(key = lambda i: np.absolute(f[i]))
 
    t = np.arange(0, n + n_predict)
    restored_sig = np.zeros(t.size)
    for i in indexes[:1 + n_harm * 2]:
        ampli = np.absolute(x_freqdom[i]) / n   # amplitude
        phase = np.angle(x_freqdom[i])          # phase
        restored_sig += ampli * np.cos(2 * np.pi * f[i] * t + phase)
    return restored_sig + p[0] * t

def from_csv(path):
    return pd.read_csv(path)


df = from_csv('vechain.csv')
x = np.asarray(df['close'])
x = x[::-1]
actual = x[-30:]
x = x[:-30]
n = 30
r = []
for i in range(len(actual)):
    r.append(len(x) + i)
extrapolation = fourierExtrapolation(x, n)

plt.plot(np.arange(0, extrapolation.size), extrapolation, 'r', label = 'extrapolation')
plt.plot(np.arange(0, x.size), x, 'b', label = 'x', linewidth = 3)
plt.plot(r, actual, 'g', label = 'actual')
plt.plot()
plt.legend()
plt.show()
```

```python
import pandas as pd
from scipy.fftpack import idct, dct
import numpy as np
import matplotlib.pyplot as plt

def from_csv(path):
    return pd.read_csv(path)

df = from_csv('vechain.csv')
l = []
l.append(np.round(df['close'].rolling(window = 5, center = False).mean(), 2))

x = []
ckD =  dct(df['close'], norm = 'ortho')
ckD[round(.9 * len(ckD)):] = 0
smoothF = idct(ckD, norm = 'ortho')
x.append(smoothF.tolist()) 

plt.scatter([i for i in range(len(x[0]))], x)
plt.scatter([i for i in range(len(x[0]))], df['close'])
#plt.scatter([i for i in range(len(l[0]))], l)
plt.show()
```
