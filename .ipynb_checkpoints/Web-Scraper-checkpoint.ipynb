```python
from bs4 import BeautifulSoup
import urllib
import datetime
import pandas as pd
import string
import urllib.request
# Get a <tr> tag with a specific class
#soup.find('tr', {'class': 'some-css-class'})

# Get element's specific attribute value
#soup.find('a', {'class': 'some-css-class'}).get('href')

# Find image's src attribute for example
#soup.find('img', {'class': 'img-responsive'}).get('src')

# Check if element has some attribute (returns True or False)
#soup.find('div').has_attr('style') 
def clean_data(self, data):
        data = data.fillna(method = 'ffill')
        data = data.fillna(method = 'bfill')
        data = data.groupby('datetime').mean()
        #for ind, rows in data.iterrows():
        #    price1 = rows['open']
        #    price2 = rows['close']
            #print("{:.2f} {:.2f}".format(price1, price2))
        return data
            
def strp(s):
    exclude = set(string.punctuation)
    x = ''.join(ch for ch in s if ch not in exclude)
    return datetime.datetime.strptime(x, '%b %d %Y')
def strp2(s):
    return s.strftime("%D")
def name_from_url(url):
    c = 0
    i = 0
    txt = ''
    while(c != '/'):
        c = url[37 + i]
        i += 1
        txt += c
    txt = txt[:-1]
    return txt

def scrape(url, dfx = None, alt_urls = None):
    b = -1
    run = True
    for i in range(len(alt_urls)+1):
        try:
            if b < 0:
                r = urllib.request.urlopen(url).read()
            else:
                r = urllib.request.urlopen(alt_urls[b]).read()
            b+=1
            run = True
            
        except urllib.error.HTTPError as err:
            run = False
            continue
            
        break
    if run:
        soup = BeautifulSoup(r, 'lxml')
        entries = []
        rows = soup.find_all("tr") # returns an array of <tr> objects
        for row in rows:
            cols = row.find_all("td")
            if len(cols) > 0:
                entry = (cols[0].get_text(), float(cols[1].get_text()), float(cols[4].get_text()))
                #time, open, close
                entries.append(entry)
        entries =  list(zip(*entries))
        entries_dict = {"time": list(entries[0]), "open": list(entries[1]), "close": list(entries[2])}
        header_list = entries_dict.keys()
        df = pd.DataFrame(columns = header_list)
        df = df.append(entries_dict, ignore_index= True)
        if dfx is not None:
            new_c = list(set(df) & set(dfx))
            df2 = df[new_c]
            dfx2 = dfx[new_c]
            df3 = pd.concat([df2, dfx2])
            df3.to_csv("crypto.csv")
        else:
            df.to_csv("crypto.csv")
    
def normalize(series, first):
    return 1.0 * series / first
#20171226
#20180326
def url_make(company, start, end):
    return 'https://coinmarketcap.com/currencies/' + company + '/historical-data/?start=' + start + '&end=' + end

def condense(df):
    for column in df:
        l = df[column].values
        print(df)
        print(l)
        print(df[column][0])
        df.loc[column] = l
        df[column].drop([i for i in range(1, len(df.index) + 1)])
    df = df.transpose()
    return df
def set_baseline(baseline, relative): # feed in as np arrays
        # Print out normalized prices
        #relative = np.asarray(relative)
        #baseline = np.asarray(baseline)
        relative = 1.0 * (relative - baseline) / baseline
        baseline = np.zeros([baseline.shape[0]])
        #baseline = baseline.tolist()
        return relative
        #pass baseline in by value??
        
def compute_correlations(self, eth, golem, neo):
        print("%.5f" % np.corrcoef(eth, golem)[0][1])
        print("%.5f" % np.corrcoef(eth, neo)[0][1])
        print("%.5f" % np.corrcoef( golem, neo)[0][1])

def load_csv(path):
    return pd.read_csv(path)

def fix_index(path, indices):
    df = pd.read_csv(path)
    df.reset_index(drop = True)
    df.index = indices[::-1]
    df.to_csv(path)
li = ['bitcoin', 'ethereum', 'ripple', 'bitcoin-cash', 'litecoin', 'eos', 'cardano', 'stellar', 'neo', 'iota', 'monero', 'dash', 'tron', 'qtum', 'vechain', 'icon', 'omisego', 'lisk', 'nano', 'zcash', 'verge', 'digixdao', 'steem', 'aeternity', 'waves', 'siacoin', 'storm', 'ontology', 'bytom', 'mithril', 'nem', 'qlink', 'metal', 'xpa', 'skycoin' ]
bi = ['20180101', '20180301','20180329']
#### MAIN #####
ind = 0
for x in li:
    url_t = url_make(x, '20171026', '20180330')
    backups = [url_make(x, i, '20180330') for i in bi]
    if ind == 0:
        scrape(url_t, alt_urls = backups)
    else:
        scrape(url_t, dfx = load_csv('crypto.csv'), alt_urls = backups)
    ind +=1
fix_index('crypto.csv', li)
#url_t = url_make('vechain', '20171026', '20180330')
#scrape(url_t)
```
