```python
import mysql.connector    
from mysql.connector import errorcode
import pandas as pd
import sqlalchemy

class DB():
    def __init__(self, file = 'Cryptodb'):
        file = open("C:/Code/db.txt", "r") 
        l = []
        for line in file: 
            l.append(line.rstrip())
        self.cnx = sqlalchemy.create_engine('mysql+mysqlconnector://{0}:{1}@{2}/{3}'.
                                                   format(l[0], l[1], l[2], 'Cryptodb'))

    def __repr__(self):
        return self.df.dropna().to_string()
        
    def to_db_from_csv(self, path):
        self.df = pd.read_csv(path)
        name = path.split('.')[0]
        self.save_db(name)
        
    def save_db(self, name_, df = None):
        
        if df is not None:
            self.df = df
        try:
            self.df.to_sql(con=self.cnx, name=name_, if_exists='replace')
            print('success')


        except mysql.connector.Error as err:
            if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
                print("Something is wrong with your user name or password")
            elif err.errno == errorcode.ER_BAD_DB_ERROR:
                print("Database does not exist")
            else:
                print(err)

    def load_db(self, table_name):
        s = 'SELECT * FROM %s' % (table_name)
        self.df = pd.read_sql(s, con=self.cnx)
```

```python
db = DB()
```

```python
db.to_db_from_csv('crypto.csv')
```
